```python
# We have four dataset:
# 1. Original dataset
# 2. Symptom-severity dataset
# 3. Symptoms_Description dataset
# 4. Symptom_precaution dataset


# Here we import some required library



import numpy as np
import pandas as pd
```

```python
# Here we load our main dataset using pandas lib


dataset=pd.read_csv('/content/dataset.csv')
dataset
```

```python
weight_dataset=pd.read_csv('/content/Symptom-severity.csv')
weight_dataset
```

```python
# Here we replace the NaN value with 0


dataset.fillna(0,inplace=True)
```

```python
dataset
```

```python
# here we convert our weight_dataset into a python dictionary so that we can replace the symptoms with there repective weights usnig a for loop


weight_dataset_dict=dict(zip(weight_dataset['Symptom'],weight_dataset['weight']))
```

```python
# List all Symptom columns
symptom_cols = [col for col in dataset.columns if col.startswith('Symptom')]

# Replace symptom strings with weights
for col in symptom_cols:
    dataset[col] = dataset[col].apply(lambda x: weight_dataset_dict.get(str(x).strip(), 0) if pd.notnull(x) else 0)
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
# Here we list all the unique values in the 'Disease' columns to convert them into numerical valuse using Encoding


dataset['Disease'].unique()
```

```python
# Here we perform encoding over the Disease column of the dataset using the LabelEncoder


from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['Disease']=le.fit_transform(dataset['Disease'])
```

```python
dataset
```

```python
# Here we separate the dataset into input and output column and also perfom the train_test_split
```

```python
x=dataset.drop(['Disease'],axis=1)
y=dataset['Disease']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
# Here we train two different models
# 1. RandomForestClassifier
# 2. AdaBoostClassifier



from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
rf_accuracy=rf.score(x_test,y_test)
print("randomforest_accuracy",rf_accuracy)

base_classifier=DecisionTreeClassifier()
adaboost_classifier=AdaBoostClassifier(base_classifier,n_estimators=50,learning_rate=0.1,random_state=42)
adaboost_classifier.fit(x_train,y_train)
adaboost_accuracy=adaboost_classifier.score(x_test,y_test)
print("adaboost_Accuracy",adaboost_accuracy)
```

```python
print(rf.score(x_train, y_train))         # For Random Forest
print(adaboost_classifier.score(x_train, y_train))  # For AdaBoost
```

```python
# Here we evaluate our model by f1_score and accuracy_score on the test dataset


from sklearn.metrics import f1_score, accuracy_score
rf_pred=rf.predict(x_test)
adaboost_pred=adaboost_classifier.predict(x_test)
print('F1-score_randomforest =', f1_score(y_test, rf_pred, average='macro')*100, '|', 'Accuracy% =', accuracy_score(y_test, rf_pred)*100)

print('F1-score_adaboost =', f1_score(y_test,adaboost_pred, average='macro')*100, '|', 'Accuracy% =', accuracy_score(y_test, adaboost_pred)*100)
```

```python
# Here we perform the Ensemble Learning using base models
# Ensemble Learning is a machine learning technique that combines predictions from multiple models to improve overall accuracy and performance compared to individual models.
# Here we use Voting Ensemble Learning



from sklearn.ensemble import VotingClassifier
ensemble_model = VotingClassifier(estimators=[
    ('randomforest', rf),
    ('adaboost', adaboost_classifier)
], voting='soft')
ensemble_model.fit(x_train, y_train)
```

```python
# Here we take input from the user
# We take only 5 input and remaining 12 places filled with zero so that input shape will remain equal


all_symptoms=list(weight_dataset_dict.keys())
print("Please select 5 symptoms from the following list:\n")
print(", ".join(all_symptoms))

selected_sym=[]
while len(selected_sym)<5:
  symptom=input(f"enter symptom {len(selected_sym)+1} :" ).strip()
  if symptom in all_symptoms and symptom not in selected_sym:
    selected_sym.append(symptom)
  else:
    print("Invalid Symptom or selected symptoms is not present in the Symptoms dictionary ")
```

```python
# Here we convert the input symptoms with there respective weights from the weight_dataset_dict


input_vector=[]
for i in selected_sym:
  input_vector.append(weight_dataset_dict.get(i,0))
```

```python
input_vector
```

```python
# Here we append zeros to the input to make the input of same size (input shape=1*17)


for i in range(17):
  if len(input_vector)<17:
    input_vector.append(0)
```

```python
input_vector
```

```python
# Convert the input into a dataframe


input_df=pd.DataFrame([input_vector],columns=["Symptom_1","Symptom_2","Symptom_3","Symptom_4","Symptom_5","Symptom_6","Symptom_7","Symptom_8","Symptom_9","Symptom_10","Symptom_11","Symptom_12","Symptom_13","Symptom_14","Symptom_15","Symptom_16","Symptom_17"])
```

```python
prediction = ensemble_model.predict(input_df)[0]
predicted_disease=le.inverse_transform([prediction])
print(f"Predicted Disease is {predicted_disease}")
```

```python
precaution_dataset=pd.read_csv("/content/symptom_precaution.csv")
precaution_dataset.head()
```

```python
# Here we convert the Precaution dataset into a dict where the key is Disease and the 4 precautions as a list as value of those keys
# Firstly set the Disease as key then transpose the dataset and covert the precaution into list and set as value to the keys

precaution_dict=precaution_dataset.set_index('Disease').T.to_dict('List')
precaution_dict
```

```python
# Here we enumerate over the precaution_dict and aquire all the precaution
#The enumerate() function in Python adds a counter to an iterable and returns it as an enumerate object


def Precaution():
  print(f"\nPrecautions for {predicted_disease_str}:")
  for i,j in enumerate(precaution_dict[predicted_disease_str],1):
      print(f"{i}. {j}")
Precaution()
```

```python
# Here we download or save all the important file which are required in prediction of disease like the model, important dictionary, and the encoder


import joblib, json
joblib.dump(ensemble_model, 'ensemble_model.pkl')
desp_file=open("desp_dataset_dict.json","w")
json.dump(desp_dataset_dict,desp_file)
desp_file.close()
precaution_file=open("precaution_dict.json","w")
json.dump(precaution_dict,precaution_file)
precaution_file.close()
weight_file=open("disease_weight_dict.json","w")
json.dump(weight_dataset_dict,weight_file)
weight_file.close()
joblib.dump(le,"_disease_Lebel_encoder")

```
